## Visualizations

```python
sns.pairplot(iris, hue='species', height=3, aspect=1)
```

```python
# Hist for all catergories combined
iris.hist(color='brown', edgecolor='black', linewidth=1.2, figsize=(12, 8))
```

```python
# Violin plots for each category separately

# Set the background color
sns.set_style(rc = {'axes.facecolor': '#e6eeff'})

plt.figure(figsize=(12, 8))
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris)
```

```python
iris.boxplot(by='species', figsize=(12, 8))
```

```python
# Scatter plot
pd.plotting.scatter_matrix(iris, figsize=(12, 10))
plt.show()
```
